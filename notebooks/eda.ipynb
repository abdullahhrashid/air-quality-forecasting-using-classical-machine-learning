```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#we will be exploring the pm25 data in this section
df = pd.read_csv('../data/interim/pm25_for_eda.csv', parse_dates = ['datetime'])
```

```python
#basic stats about pm2.5 levels for the 7 years
df.describe()
```

```python
#checking the amount of nulls
df.isnull().sum()
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#plotting and visualizng
plt.plot(df['datetime'], df['value'])
plt.xlabel('Time', weight = 'bold')
plt.ylabel('PM2.5 Levels', weight = 'bold')
plt.title('PM2.5 Data', weight = 'bold')
plt.grid(alpha = 0.3, linestyle = '--', color = 'black')
plt.show()
```

```python
#processed dataset after interpolation
df = pd.read_csv('../data/processed/pm25.csv', parse_dates = ['datetime'])
```

```python
#basic stats after interpolation
df.describe()
```

```python
#after interpolation plot
plt.plot(df['datetime'], df['value'], c = 'maroon')
plt.xlabel('Time', weight = 'bold')
plt.ylabel('PM2.5 Levels', weight = 'bold')
plt.title('PM2.5 After Interpolation', weight = 'bold')
plt.grid(alpha = 0.3, linestyle = '--', color = 'black')
plt.show()
```

```python
#analyzing the main dataset now
df = pd.read_csv('../data/lagged_dataset.csv')
```

```python
import seaborn as sns

#removing categorical features for heatmap
df = df.drop(columns = ['datetime', 'conditions'])
corr_matrix = df.corr()

#plotting correlation matrix as heatmap
plt.figure(figsize = (15,15))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
